```python
df = pd.read_csv("/kaggle/input/datasets/shayanfazeli/heartbeat/ptbdb_normal.csv", header = None)
print(df[187].value_counts())

print(df.shape)
print(df.head())
```

Visualizing the data above

```python
import matplotlib.pyplot as plt

# Plot the first heartbeat
plt.plot(df.iloc[0, :187])
plt.title('One normal heartbeat')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

# Numpy core application

```python
plt.plot(df.iloc[0,25:100])#slicing is donr here
plt.title('One normal heartbeat')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python
#aggregate part
print("----------STATS----------")
print("avg heartbeat: ", np.mean(df))
print("std deviation: ", np.std(df))
print("variance: ", np.var(df))
print("Min beat: ", np.min(df))
print("Max beat: ", np.max(df))
```

# Matplotlib basics application
