# Tensors

```python
x = torch.tensor([2.5,0.1])
print(x)
```

```python
x = torch.rand(2,2)
y = torch.rand(2,2)
print(x)
print(y)
z=x+y
z = torch.add(x,y)
print(z)
```

```python
x = torch.rand(2,2)
y = torch.rand(2,2)
y.add_(x) #trailing _ will modify the value
print(y)
```

```python
x = torch.rand(2,2)
y = torch.rand(2,2)
z=x*y
z=torch.mul(x,y)
y.mul_(x)
```

```python
x = torch.rand(5,3)
print(x)
print(x[:, 0])
print(x[1,:])
print(x[1, 1])
print(x[1, 1].item()) #get's actual value but only works with one value in there
```

```python
x=torch.rand(4,4)
print(x)
y=x.view(16)
print(y.size())
```

```python
a = torch.ones(5)
print(a)
b=a.numpy()
print(b)

a.add_(1)
print(a)
print(b) #same memory location
```

```python
a = np.ones(5)
print(a)
b = torch.from_numpy(a)
print(b)
```

```python
a+=1
print(a)
print(b)
```

```python
if torch.cuda.is_available(): #for gpu instead of cpu
    device = torch.device("cuda")
    x = torch.ones(5, device=device)
    y = torch.ones(5)
    y = y.to(device)
    z = x+y
    z.to("cpu") #to get back to cpu
print(z)
```

```python
x = torch.ones(5, requires_grad = True)
print(x)
```
